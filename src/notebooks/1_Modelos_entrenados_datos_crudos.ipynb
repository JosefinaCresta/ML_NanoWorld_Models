```python
def del_zeros(df):
    precent_zeros = (df==0).sum()*100/len(df)
    zeros_value_df = pd.DataFrame({'column_name': df.columns,
                                'percent_zeros': precent_zeros}).sort_values('percent_zeros', ascending=False)
    cols_to_drop = zeros_value_df[zeros_value_df['percent_zeros'] > 99.99].index.values
    print("Cols:", cols_to_drop)

    print("Columnas pre drop:", len(data.columns))

    data.drop(columns=cols_to_drop, inplace=True)

    print("Columnas post drop:", len(data.columns))
```

```python
def modeling_scoring(model, parameters, df, target):
    print("model: ", model)
    print("Parametros ", model.get_params())
    X = data.drop([target], 1)
    y = data[[target]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
    
    modelo_grid = GridSearchCV(estimator= model , 
    param_grid = parameters,
    scoring = "neg_root_mean_squared_error", 
    cv = 10)

    mejor_modelo = modelo_grid.fit(X_train, y_train)
    # print("best_estimator_.get_params: ", mejor_modelo.best_estimator_.get_params())
    y_pred = mejor_modelo.predict(X_test)

    print('Best Hyperparameters: %s' % mejor_modelo.best_params_)
    print('Best Score: %s' % mejor_modelo.best_score_)

    # score = rmsle_cv(lr)
    # print("\nJuan score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

    print("Score del modelo (R^2):", round(lr.score(X_test, y_test), 4))
    print("R^2 score:", round(r2_score(y_test, y_pred), 4))
    print("MAE score:", round(mean_absolute_error(y_test, y_pred), 4))
    print("MSE score:", round(mean_squared_error(y_test, y_pred), 4))
    print("RMSE score:", round(np.sqrt(mean_squared_error(y_test, y_pred)), 4))
    print("RMSLE score:", round(np.sqrt(mean_squared_log_error(y_test, y_pred)), 4))

    def mean_absolute_percentage_error(y_pred, y_true): 
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


    print("MAPE score:", round(mean_absolute_percentage_error(y_test, y_pred), 4))
```
