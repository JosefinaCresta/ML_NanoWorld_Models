# LinearRegression()

Gráfica de residuos

Los residuos, en el contexto de los modelos de regresión, son la diferencia entre el valor observado de la variable objetivo (y) y el valor predicho (ŷ), es decir, el error de la predicción. El gráfico de residuales muestra la diferencia entre los residuales en el eje vertical y la variable dependiente en el eje horizontal, lo que le permite detectar regiones dentro del objetivo que pueden ser susceptibles a más o menos errores.
Un uso común de la gráfica de residuos es analizar la varianza del error del regresor. Si los puntos están dispersos aleatoriamente alrededor del eje horizontal, un modelo de regresión lineal suele ser apropiado para los datos; de lo contrario, un modelo no lineal es más apropiado. En el caso anterior, vemos una distribución uniforme y bastante aleatoria de los residuos contra el objetivo en dos dimensiones. Esto parece indicar que nuestro modelo lineal está funcionando bien. También podemos ver en el histograma que nuestro error normalmente se distribuye alrededor de cero, lo que generalmente también indica un modelo bien ajustado.

Tenga en cuenta que si no se desea el histograma, se puede desactivar con el indicador hist=False:
